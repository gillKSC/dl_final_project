```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('train_acc_epoch.pickle', 'rb') as f:
    loaded_data = pickle.load(f)
    
plt.plot(loaded_data)
plt.xlabel('epoch')
plt.ylabel('training')
```

```python
with open('train_acc_batch.pickle', 'rb') as f:
    loaded_data = pickle.load(f)
    
plt.plot(loaded_data)
plt.ylabel('training')
plt.xlabel('batch')
```
